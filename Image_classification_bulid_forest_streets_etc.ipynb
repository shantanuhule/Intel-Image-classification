```python
import pandas as pd
import numpy as np
import cv2
from tqdm import  tqdm 
import tensorflow as tf
import os
import matplotlib.pyplot as plt
from random import shuffle
```

```python
IMG_SIZE=150
training_data=[]
testing_data=[]
```

```python
#training data
build_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\buildings'
forest_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\forest'
glacier_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\glacier'
mountain_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\mountain'
sea_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\sea'
street_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_train\seg_train\street'

#labels 
#  1 - bulidings
#  2 - forest
#  3 - glacier
#  4 - mountain
#  5 - sea
#  6 - street
def train_data():
    for img in tqdm(os.listdir(build_dir)):
        label=1
        path=os.path.join(build_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))

        training_data.append([np.array(img_rs),np.array(label)])
        
    for img in tqdm(os.listdir(forest_dir)):
        label=2
        path=os.path.join(forest_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        training_data.append([np.array(img_rs),np.array(label)])
        
    
    for img in tqdm(os.listdir(glacier_dir)):
        label=3
        path=os.path.join(glacier_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        training_data.append([np.array(img_rs),np.array(label)])


    for img in tqdm(os.listdir(mountain_dir)):
        label=4
        path=os.path.join(mountain_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        training_data.append([np.array(img_rs),np.array(label)])        

        
    for img in tqdm(os.listdir(sea_dir)):
        label=5
        path=os.path.join(sea_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if(type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        training_data.append([np.array(img_rs),np.array(label)])
        
    for img in tqdm(os.listdir(street_dir)):
        label=6
        path=os.path.join(street_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        training_data.append([np.array(img_rs),np.array(label)])
        
        

        
```

```python
#testing data
build_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\buildings'
forest_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\forest'
glacier_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\glacier'
mountain_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\mountain'
sea_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\sea'
street_dir=r'C:\Users\Comp\Desktop\Deep Learning\Convolutional Neural Networks\Convolution Neural Network\img_classification\seg_test\seg_test\street'
#labels 
#  1 - bulidings
#  2 - forest
#  3 - glacier
#  4 - mountain
#  5 - sea
#  6 - street
def test_data():
    for img in tqdm(os.listdir(build_dir)):
        label=1
        path=os.path.join(build_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))

        testing_data.append([np.array(img_rs),np.array(label)])
        
    for img in tqdm(os.listdir(forest_dir)):
        label=2
        path=os.path.join(forest_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_rs),np.array(label)])
        
    
    for img in tqdm(os.listdir(glacier_dir)):
        label=3
        path=os.path.join(glacier_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_rs),np.array(label)])


    for img in tqdm(os.listdir(mountain_dir)):
        label=4
        path=os.path.join(mountain_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_rs),np.array(label)])        

        
    for img in tqdm(os.listdir(sea_dir)):
        label=5
        path=os.path.join(sea_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if(type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_rs),np.array(label)])
        
    for img in tqdm(os.listdir(street_dir)):
        label=6
        path=os.path.join(street_dir,img)
        img=cv2.imread(path,cv2.IMREAD_COLOR)
        if (type(img))==type(None):
            pass
        else:
            img_rs=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_rs),np.array(label)])
        
        

        
```

```python
print("lenght of training data:",len(training_data))
print("length of testing data:",len(testing_data))
```

```python
X_train=np.array([i[0] for i in training_data])
Y_train=np.array([i[1] for i in training_data])

X_test=np.array([i[0] for i in testing_data])
Y_test=np.array([i[1] for i in testing_data])


print("X_train shape:",X_train.shape)
print("Y_train shape:",Y_train.shape)
print("X_test shape:",X_test.shape)
print("Y_test shape:",Y_test.shape)
```

```python
# normalization 
X_train=X_train/255
X_test=X_test/255
```

```python
#one hot

Y_train=tf.one_hot(Y_train,6)
Y_test=tf.one_hot(Y_test,6)

print(Y_train[0])
print("Y_train shape:",Y_train.shape)
```

```python
#display image
index=15
print(np.argmax(Y_train[index]))
plt.imshow(X_train[index])
plt.show()
```

```python
def model_classification():
    
    model=tf.keras.Sequential([ tf.keras.layers.InputLayer((IMG_SIZE,IMG_SIZE,3)),
                              tf.keras.layers.ZeroPadding2D(padding=(3,3)),
                              tf.keras.layers.Conv2D(filters=32,kernel_size=7,strides=(1,1),activation='relu'),
                              tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001,beta_initializer='zeros', gamma_initializer='ones',moving_mean_initializer='zeros',moving_variance_initializer='ones'),
                              #tf.keras.layers.ReLU(),
                              tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid'),
                              #tf.keras.layers.Dropout(0.25),
                               
                              tf.keras.layers.ZeroPadding2D(padding=(3,3)),
                              tf.keras.layers.Conv2D(filters=32,kernel_size=7,strides=(1,1),activation='relu'),
                              tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001,beta_initializer='zeros', gamma_initializer='ones',moving_mean_initializer='zeros',moving_variance_initializer='ones'),
                              #tf.keras.layers.ReLU(),
                              tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid'),
                              #tf.keras.layers.Dropout(0.25),
                               
                              tf.keras.layers.ZeroPadding2D(padding=(3,3)),
                              tf.keras.layers.Conv2D(filters=32,kernel_size=7,strides=(1,1),activation='relu'),
                              tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001,beta_initializer='zeros', gamma_initializer='ones',moving_mean_initializer='zeros',moving_variance_initializer='ones'),
                              #tf.keras.layers.ReLU(),
                              tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid'),
                              #tf.keras.layers.Dropout(0.25),
                               
                              tf.keras.layers.ZeroPadding2D(padding=(3,3)),
                              tf.keras.layers.Conv2D(filters=32,kernel_size=7,strides=(1,1),activation='relu'),
                              tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001,beta_initializer='zeros', gamma_initializer='ones',moving_mean_initializer='zeros',moving_variance_initializer='ones'),
                              #tf.keras.layers.ReLU(),
                              tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid'),
                              #tf.keras.layers.Dropout(0.25),
                               
                              tf.keras.layers.ZeroPadding2D(padding=(3,3)),
                              tf.keras.layers.Conv2D(filters=32,kernel_size=7,strides=(1,1),activation='relu'),
                              tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001,beta_initializer='zeros', gamma_initializer='ones',moving_mean_initializer='zeros',moving_variance_initializer='ones'),
                              #tf.keras.layers.ReLU(),
                              tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid'),
                              #tf.keras.layers.Dropout(0.25),
                               
                              tf.keras.layers.Flatten(),
                              tf.keras.layers.Dense(6,activation='softmax')
                              ])
    return model
```

```python
model=model_classification()
```

```python
model.compile(optimizer='adam', loss='CategoricalCrossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=16)
```

```python
model.evaluate(X_test,Y_test)
```

```python
test_img=X_test[0]
test_img=(np.expand_dims(test_img,0))
print(model.predict(test_img))
print("class predicted:",np.argmax(model.predict(test_img)))
```
